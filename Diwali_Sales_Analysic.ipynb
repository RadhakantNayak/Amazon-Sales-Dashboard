From The above we got That Column Status and Unnamed1 is empty And In Amount Column There some Null Values

```python
df.drop(["Status","unnamed1"],axis = 1,inplace = True)
df.head(10)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True , ignore_index = True)   
# Here We drop The null Value and Reset The index again
```

```python
df
```

```python
a = df.duplicated(keep = False)   
df[a]   # From The Above We Get All The Duplicated values So We are Going To Drop The Duplicated Values
```

```python
df.drop_duplicates(inplace = True,ignore_index = True)
```

```python
df
```

```python
df_copy = df.copy()  # I have copied the column and saved into another variable for future require
df_copy
```

```python
df_copy["Total_Amount"] = df_copy["Orders"] * df_copy["Amount"]
df_copy
```

# Exploratory Data Analysic

###  1. Gender wise purchase Quantity And  Spending Amount


```python
df_copy.head(4)
```

```python
 ax = sns.countplot(x = "Gender",data = df_copy)
    
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_wise_total = df_copy.groupby('Gender')['Total_Amount'].sum()



sns.barplot(x=gender_wise_total.index, y=gender_wise_total.values)
plt.xlabel('Gender')
plt.ylabel('Total Sale')
plt.title('Gender-wise Total Sale')
plt.show()
```

#### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than male

##  2. Which age group contributing more price


```python
Group_Age = df_copy[["Age Group","Total_Amount"]]
Group_Age.head()
```

```python
ay = sns.histplot(x = "Age Group",data = Group_Age )

for bars in ay.containers:
    ay.bar_label(bars)
```

```python
age_group_groupby = Group_Age.groupby(["Age Group"])["Total_Amount"].sum()

sortedv = age_group_groupby.sort_values(ascending = False)

sortedv
```

```python
sns.barplot(x = sortedv.index, y = sortedv.values)
```

```python

### From Aboe We Found That 26-35 Age Group Creating More Sale
```

## 3 . Which State Contributing More Sale

```python
df_copy.head()
```

```python
sales_state = df_copy.groupby(['State'])['Orders'].sum()
Sort_Sales = sales_state.sort_values(ascending = False).head(8)
```

```python
ax = sns.barplot(x = Sort_Sales.index,y = Sort_Sales.values)
plt.figure(figsize = (16,5))


for bars in ax.containers:
    ax.bar_label(bars)
```

#### From Uttar Pradesh State we got Highest Number of count Orders 4803

```python
state_amt = df_copy.groupby('State')['Total_Amount'].sum()
Sort_State_amt = state_amt.sort_values(ascending = False).head(8)



ax = sns.barplot(x=Sort_State_amt.index, y=Sort_State_amt.values)
plt.xlabel('State')
plt.ylabel('Total Sale')
plt.title('State-wise Total Sale')
plt.show()

```

##  4. Zone Wise Total Sale

```python
df_Zone = df_copy.groupby(["Zone"])["Total_Amount"].sum()

plt.figure(figsize=(5, 5))
plt.pie(x=df_Zone.values, labels=df_Zone.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Create a donut chart by adding a circle at the center
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')  

plt.show()
```

#### Central Zone Is Highest Total Of 39%

## 5. Top 5 Product Category on Basic Of Total Sale.

```python
df_cat = df_copy.groupby(["Product_Category"])["Total_Amount"].sum()
df_sort = df_cat.sort_values(ascending = False).head(5)
df_sort
```

```python
ax = sns.barplot(x=df_sort.index, y=df_sort.values)
plt.xlabel('Product Category')
plt.ylabel('Total Sale')
plt.title('Category Total Sale')
plt.show()
```

#### Food is the highest ordered category
